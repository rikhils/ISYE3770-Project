### Question 1: What are the overall trends in the dataâ€™s distributions? 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from cci import charlson_comorbidity_idx
import json


# Load the data
data = pd.read_json('cci_kits23.json')

# data = data.dropna()
```

```python
# Age Distribution

# Mean age
mean_age = np.mean(data['age_at_nephrectomy'])
print('Mean age:', mean_age)
# Median age
median_age = np.median(data['age_at_nephrectomy'])
print('Median age:', median_age)
# Standard deviation
std_age = np.std(data['age_at_nephrectomy'])
print('Standard deviation:', std_age)

plt.figure(figsize=(10, 6))
sns.displot(data['age_at_nephrectomy'], bins=30, kde=False)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Gender Demographics

# Gender distribution
male_count = data[data["gender"] == "male"].shape[0]
print("Number of males:", male_count)

female_count = data[data["gender"] == "female"].shape[0]
print("Number of females:", female_count)
```

```python
# Distribution of Tumor Stage

data['pathology_t_stage'] = data['pathology_t_stage'].apply(lambda x: '1' if x in ['1a', '1b'] else ('2' if x in ['2a', '2b'] else x))
order = ['na', '1', '2', '3', '4']
 
plt.figure(figsize=(10, 6))
sns.countplot(x = data['pathology_t_stage'], data=data, order=order)
plt.title('Distribution of Tumor Stage')
plt.xlabel('Tumor Stage')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Tumor ISUP Grade

plt.figure(figsize=(10, 6))
order = ['na', '1', '2', '3', '4']
sns.countplot(x = 'tumor_isup_grade', data=data, order=order)    
plt.title('Distribution of Tumor ISUP Grade')
plt.xlabel('Tumor ISUP Grade')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Surgical Procedure

plt.figure(figsize=(10, 6))
# order = ['na', '1', '2', '3', '4']
sns.countplot(x = 'surgical_procedure', data=data)    
plt.title('Distribution of Surgical Procedures')
plt.xlabel('Surgical Procedure')
plt.ylabel('Count')
plt.show()

# Distribution of Surgery Type

plt.figure(figsize=(10, 6))
# order = ['na', '1', '2', '3', '4']
sns.countplot(x = 'surgery_type', data=data)    
plt.title('Distribution of Surgery Types')
plt.xlabel('Surgery Type')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of BMI

# Mean BMI
mean_bmi = np.mean(data['bmi'])
print('Mean BMI:', mean_bmi)
# Standard deviation
std_bmi = np.std(data['bmi'])   
print('Standard deviation:', std_bmi)

plt.figure(figsize=(10, 6))
sns.displot(data['bmi'], bins=30, kde=False)
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Smoking Pack Years

# Mean Smoking Pack Years
mean_pack_years = np.mean(data['pack_years'])
print('Mean Smoking Pack Years:', mean_pack_years)
# Standard deviation
std_pack_years = np.std(data['pack_years'])
print('Standard deviation:', std_pack_years)

plt.figure(figsize=(10, 6))
sns.displot(data['pack_years'], bins=30, kde=False)
plt.title('Smoking Pack Years Distribution')
plt.xlabel('Smoking Pack Years')
plt.ylabel('Count')
plt.show()
```

```python
# Charleson Comorbidity Index

# Mean CCI Points
mean_cci = np.mean(data['cci_points'])
print('Mean CCI Points:', mean_cci)
# Standard deviation
std_cci = np.std(data['cci_points'])
print('Standard deviation:', std_cci)
# Median CCI Points
median_cci = np.median(data['cci_points'])
print('Median CCI Points:', median_cci)

plt.figure(figsize=(10, 6))
sns.countplot(x = 'cci_points', data=data)
plt.title('Distribution of Charleson Comorbidity Index (CCI) Points')
plt.xlabel('CCI Points')
plt.ylabel('Count')
plt.show()

# data['cci_ten_year_survival'] = data['cci_ten_year_survival'].apply(lambda x: x*100)

# Mean CCI 10 year survival
mean_cci_survival = np.mean(data['cci_ten_year_survival'])
print('Mean CCI 10 year survival:', mean_cci_survival)
# Standard deviation
std_cci_survival = np.std(data['cci_ten_year_survival'])
print('Standard deviation:', std_cci_survival)


plt.figure(figsize=(10, 6))
sns.displot(data['cci_ten_year_survival'], bins=100, kde=False)
plt.title('Charleson Comorbidity Index (CCI) 10 year survival Distribution')
plt.xlabel('% Chance of 10 year survival')
plt.ylabel('Count')
plt.show()

```
